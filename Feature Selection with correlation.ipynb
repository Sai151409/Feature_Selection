#### 2. Feature Selection- With Correlation
In this step we will be removing the features which are highly correlated

```python
from sklearn.datasets import load_boston
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')
```

```python
data = load_boston()
df = pd.DataFrame(data.data, columns = data.feature_names)
df['MEDV'] = data.target
```

```python
df.head()
```

```python
df.corr()
```

```python
df.corr()
```

```python
plt.figure(figsize=(10, 10))
sns.heatmap(df.corr(), annot = True)
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(df.drop(columns='MEDV', axis = 1), df['MEDV'], test_size=0.3)
```

```python
def high_corr(corr_matrix, threshold):
    columns = set()
    for i in range(len(corr_matrix.columns)):
        for j in range(i):
            if abs(corr_matrix.iloc[i, j]) > threshold:
                columns.add(corr_matrix.columns[i])
    return columns
```

```python
multi_collinearity = list(high_corr(X_train.corr(), 0.7))
```

```python
multi_collinearity
```

```python
X_train.drop(columns=multi_collinearity, axis = 1)
```
